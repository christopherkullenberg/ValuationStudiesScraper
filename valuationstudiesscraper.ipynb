```python
import requests
import io
import re
from bs4 import BeautifulSoup
```

```python
vsurlfile = open("valuationstudiesallURLs.txt")
vsurls = vsurlfile.readlines()

allVSdoifile = open("VSdoi.txt", "w") # The DOIs of each VS article
```

```python
for v in vsurls:
    doi = re.findall(r'DOI\=(.*)', v, re.IGNORECASE)
    print(v)
    print(doi)
    allVSdoifile.write(doi[0] + "\n") # Write each VS-articles own DOI to file
allVSdoifile.close()
```

```python
VSreferencesDOIs = open("VSreferencesDOIs.txt", "w")
```

```python
counter = 0


for url in vsurls:
    print("Working with " + url)
    r = requests.get(url)
    file_like_obj = io.StringIO(r.text) #Turns the requested output into a file like objet
    lines = file_like_obj.read()
    counter += 1 #Update the counter from proper filenames
    soup = BeautifulSoup(lines, "lxml")
    #print(soup)
    hrefs = soup.find_all('a', href=True) #This retrieves each url
    #print(hrefs)
    for h in hrefs:
        if h['href'].startswith("https://doi.org/10.3384/VS"): #clean out self-citations
            continue
        elif h['href'].startswith("https://doi.org/10.3384/vs"): #clean out self-citations
            continue
        elif h['href'].startswith("https://doi.org/"):
            print(h['href'][16:])
            VSreferencesDOIs.write(h['href'][16:] + "\n")

VSreferencesDOIs.close()


```
